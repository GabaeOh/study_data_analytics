#### quest 2
- 직업군이 상위 1, 2위 항목 : value_count, iloc
- 체중 > 50, 신장 > 180, 흡연여부 없는 분
- 해당 레코드 갯수 표현
- option : value_count 결과 변수로 활용

```python
df_ROS_forpatient[['직업']].value_counts() # 직업군이 상위 1,2위에 속하는 항목 확인
```

```python
df_ROS_forpatient['직업'].value_counts().index
```

```python
df_ROS_forpatient['직업'].value_counts().index[0]
```

```python
df_df = df_ROS_forpatient['직업'].value_counts().iloc[:2]
df_df
```

```python
df_df.index[1]
```

```python
condition = (df_ROS_forpatient['직업'] == '자영업') | (df_ROS_forpatient['직업'] == '사무직')  #'자영업', '사무직' 항목만 추출 (하드코딩 방법)
df_ROS_QUEST2 = df_ROS_forpatient.loc[condition] 
df_ROS_QUEST2
```

```python
condition = (df_ROS_forpatient['직업'] == df_df.index[1]) | (df_ROS_forpatient['직업'] == df_df.index[0]) # 직업군 1,2위만 추출 (하드코딩 x)
df_ROS_forpatient.loc[condition]
```

```python
condition = (df_ROS_QUEST2['체중'] > 70) & (df_ROS_QUEST2['신장'] > 170) & (df_ROS_QUEST2['흡연여부'] == 0) # 조건 걸기
df_ROS_QUEST22 = df_ROS_QUEST2.loc[condition]
df_ROS_QUEST22
```

```python
df_ROS_QUEST22.shape  # 해당 레코드 갯수 표현
```

#### quest3 
- quest2 최종 데이터 사용
- 직업별 성별 빈도 : 'size'
- 직업별 혈액형 빈도 : 'size'

```python
import numpy as np
```

```python
df_ROS_QUEST22.pivot_table(index='직업', columns='성별', aggfunc='size') # 직업별 성별 빈도 1번째 방법
```

```python
pd.pivot_table(df_ROS_QUEST22[['성별', '직업']], index='성별',columns='직업', aggfunc=np.size) # 직업별 성별 빈도 2번째 방법 
```

```python
df_ROS_QUEST22.pivot_table(index='직업', columns='혈액형', aggfunc='size')  # 직업별 혈액형 빈도 1번째 방법
```

```python
pd.pivot_table(df_ROS_QUEST22[['직업', '혈액형']], index=['혈액형'],columns='직업', aggfunc=np.size) # 직업별 혈액형 빈도 2번째 방법
```
