```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
```

```python
df
```

```python
df.drop_duplicates() # 전체 컬럼 기준
```

```python
df.drop_duplicates(subset=['brand', 'style'], keep='last') # 특정 컬럼 기준 # keep='last' = 중복된 컬럼중에서 가져올 컬럼
```
